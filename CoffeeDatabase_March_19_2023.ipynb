## Data Dictionary:-  

ID = Unique identification number of the coffee report.  
Country of Origin = The country the coffee originated from.  
Farm Name = The farm where the coffee was cultivated and harvested.  
Lot Number = The processing batch number of coffee beans.  
Mill = The processing station which the coffee beans have been processed in.  
ICO Number = A combination of figures which includes the country code (3 figures), the code for the exporter or grower (4 figures) and the parcel number (4 figures). The ICO mark must be unique to each parcel of coffee.  
Company = The company that sends the coffee beans for grading.  
Altitude = The altitude (height relative to sea level) at which the coffee plants are grown.  
Region = The region where the coffee is cultivated and harvested.  
Producer = The producer of the harvested coffee fruit.  
Number of Bags = Number of bags for each Lot.  
Bag Weight = Weight of the bag in (Kilograms(kg))  
In-Country Partner = An establishment that serve as hubs to evaluate coffee quality within the Q Coffee System.  
Harvest Year = The year the coffee fruit was harvested.  
Grading Date = The date the coffee beans were graded.  
Owner = Owner of the coffee beans that were graded.  
Variety = The subspecies of the coffee plant which the coffee beans came from.  
Status = The status of the Q grading.  
Processing Method = The method in which the coffee fruit is processed.  
Aroma = The evaluation of the smell of the coffee grounds.  
Flavor = The evaluation of the taste of the coffee(the main flavours that are noticable.)  
Aftertaste = The evaluation of how long the coffee's positive flavor lingers.(the longer it is the higher the grading)  
Acidity = The evaluation of the level of acidity in the coffee.(Should not be sour or dominant flavour but should be subtle and enhances other aspects of the coffee.)  
Body = The evaluation of a coffee's texture or feel in the mouth in relation to a coffee's other qualities.  
Balance = Evaluation of how a coffee's different qualities interact with each other.  
Uniformity = Evaluation of the tasting consistency of the coffee bean test sample.  
Clean Cup = Evaluation of unusual flavours and aromas in the coffee.  
Sweetness = Evaluation of a coffee's fullness of flavor and sweet flavors.  
Overall = The evaluation of all the coffee qualities as a whole.  
Defects = Points Deducted from the Q Score depending on the type of defect.  
Total Cup Points = The overall rating of the coffee.  
Moisture = Percentage of the moisture content in the green unroasted coffee beans.  
Category One Defects = Number of Primary Defects found which include the following: (Full black,Full sour,Dried cherry/pod,Fungus damage,Foreign matter,Severe insect damage).  
Quakers = Number of immature coffee beans that are found in the sample.  
Color = The color of unroasted coffee beans.  
Category Two Defects = Number of Secondary Defects found which include the following:(Partial black,Partial sour,Parchment,Floater,Immature/unripe,Withered,Shell,Broken/chipped/cut,Hull/husk,Slight insect damage).  
Expiration = The Date that the certification of the coffee expires on.  
Certification Body = The organization which certified the coffee.  
Certification Address = The address of the organization which certified the coffee  
Certification Contact = The contact person regarding the certification of the coffee.  
Species = The species which the coffee beans belong to.
