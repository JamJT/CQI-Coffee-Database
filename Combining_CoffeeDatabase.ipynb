```python
# Importing pandas library to create and manipulate Data Frames. Editing DataFrame viewing for no limit reading.
import pandas as pd
#pd.set_option('display.max_columns',None)
#pd.set_option('display.max_colwidth',None)
#pd.set_option('display.max_rows',None)
#pd.set_option('display.width',None)
```

```python
#Function for arranging the shape and size of the tables (1-3) that are being imported
def ct(coffeetable):
    coffeetable.drop(columns=['Unnamed: 0'],inplace=True)
    table1 = coffeetable[['0','1']]
    table2 = coffeetable[['2','3']].rename(columns={'2':'0','3':'1'})
    coffeetable = pd.concat([table1,table2],ignore_index=True)
    coffeetable = coffeetable.T
    coffeetable.columns = coffeetable.iloc[0]
    coffeetable.drop( labels='0',inplace=True)
    return coffeetable
```

```python
#Function for arranging the shape and size of the tables (4) that are being imported. A separate function because Table 4 has a different shape than the other tables.
def ct4(coffeetable):
    coffeetable.drop(columns=['Unnamed: 0'],inplace=True)
    coffeetable = coffeetable.T
    coffeetable.columns = coffeetable.iloc[0]
    coffeetable.drop( labels='0',inplace=True)
    return coffeetable
```

```python
# A for Loop to apply the (ct) function for the imported datasets and then combining and storing the datasets into a singular DataFrame.
coffeedatabase = pd.DataFrame()
for i in range(0,259):
    coffeereport = pd.DataFrame()
    for j in range(1,4):
        coffeebean = pd.read_csv(f'coffee_{i}_table_{j}.csv') 
        coffeereport = pd.concat([coffeereport,ct(coffeebean)], axis = 1)
    print(coffeereport)
    coffeedatabase = pd.concat([coffeedatabase,coffeereport], axis = 0, ignore_index = True)
```
